```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.random.randint(low=-5, high=5, size=(100, 2))
w1=[];w2=[];w3=[];wir=[];
x, y = mat[:, 0], mat[:, 1]
plt.plot(x, y, 'o')
```

```python
d12 = -2*x + 3*y - 1
d13 = -10*x - y - 1
d23 = -x + y - 1

for i in range(mat.shape[0]):
    if d12[i] > 0 and d13[i] > 0:
        w1.append(mat[i].tolist())
    elif d12[i] < 0 and d23[i] > 0:
        w2.append(mat[i].tolist())
    elif d13[i] < 0 and d23[i] < 0:
        w3.append(mat[i].tolist())
    else:
        wir.append(mat[i].tolist())

x0 = np.linspace(-5, 5, 100)
y1 = -2/3 * x0 - 1
y2 = -10*x0 - 1
y3 = x0 + 1

w1 = np.array(w1)
w2 = np.array(w2)
w3 = np.array(w3)
wir = np.array(wir)

plt.plot(x0, y1, 'r')
plt.plot(x0, y2, 'g')
plt.plot(x0, y3, 'b')
if w1.size > 0: 
    plt.plot(w1[:,0],w1[:,1],'o',label='w1')
if w2.size > 0:
    plt.plot(w2[:,0],w2[:,1],'*',label='w2')
if w3.size > 0:
    plt.plot(w3[:,0],w3[:,1],'^',label='w3')
if wir.size > 0:
    plt.plot(wir[:,0],wir[:,1],'.',label='IR')
plt.axis([-5.5, 5.5, -5.5, 5.5])
plt.legend()
```

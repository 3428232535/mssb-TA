```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.random.randint(low=-10, high=10, size=(10, 2))
plt.plot(mat[:, 0], mat[:, 1], 'o')

x = np.linspace(-10, 10, 100)
y1 = x - 1
y2 = - x + 4
y3 = 0 * x + 1
plt.plot(x, y1, 'r', label='$d_1: -x+y+1=0 $')
plt.plot(x, y2, 'g', label='$d_2: x+y-4=0 $')
plt.plot(x, y3, 'b', label='$d_3: y-1=0 $')

plt.legend()



```

```python
d1 = np.frompyfunc(lambda x,y: 1 if -x+y+1 > 0 else 0, 2, 1)(mat[:, 0], mat[:, 1])
d2 = np.frompyfunc(lambda x,y: 1 if x+y-4 > 0 else 0, 2, 1)(mat[:, 0], mat[:, 1])
d3 = np.frompyfunc(lambda x,y: 1 if y-1 > 0 else 0, 2, 1)(mat[:, 0], mat[:, 1])
D = np.concatenate((d1.reshape(-1, 1), d2.reshape(-1, 1), d3.reshape(-1, 1)), axis=1)
for i in range(D.shape[0]):
    match (D[i, 0], D[i, 1], D[i, 2]):
        case (1, 0, 0):print(f'({mat[i, 0]},{mat[i, 1]}) in w1')
        case (0, 1, 0):print(f'({mat[i, 0]},{mat[i, 1]}) in w2')
        case (0, 0, 1):print(f'({mat[i, 0]},{mat[i, 1]}) in w3')
        case (1, 1, 0):print(f'({mat[i, 0]},{mat[i, 1]}) in w1/w2')
        case (1, 0, 1):print(f'({mat[i, 0]},{mat[i, 1]}) in w1/w3')
        case (0, 1, 1):print(f'({mat[i, 0]},{mat[i, 1]}) in w2/w3')
        case (1, 1, 1):print(f'({mat[i, 0]},{mat[i, 1]}) in w1/w2/w3')
        case (0, 0, 0):print(f'({mat[i, 0]},{mat[i, 1]}) not in w1/w2/w3')
```

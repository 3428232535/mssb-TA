```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  
# Matplotlib中设置字体-黑体，解决Matplotlib中文乱码问题
```

```python
def perceptron(x1,x2,w,e):
    m1, _ = x1.shape
    m2, _ = x2.shape
    X1 = np.concatenate((x1,np.ones((m1,1))),axis=1)
    X2 = np.concatenate((x2,np.ones((m2,1))),axis=1)
    X2 = -X2
    flag = False
    while not flag:
        flag = True
        for i in range(m1):
            if np.dot(w,X1[i]) <= 0:
                w += e*X1[i]
                flag = False
        for i in range(m2):
            if np.dot(w,X2[i]) <= 0:
                w += e*X2[i]
                flag = False
    return w
```

```python
x1 = np.array([
    [1.24,1.27],
    [1.36,1.74],
    [1.38,1.64],
    [1.38,1.82],
    [1.38,1.90],
    [1.40,1.70],
    [1.48,1.82],
    [1.54,1.82],
    [1.56,2.08]
])
x2 = np.array([
    [1.14,1.82],
    [1.18,1.96],
    [1.20,1.86],
    [1.26,2.00],
    [1.28,2.00],
    [1.30,1.96]
])
x3=np.array([1.40,2.04]).reshape([1,2])
x4=np.array([[1.24,1.80],[1.28,1.84]])
# 初始化并计算权向量
w=np.array([0.1,0.1,0.1]).reshape([1,3])
w=perceptron(x1,x2,w,0.1)
```

```python
plt.axis([0.9,1.7,1.1,2.2])
plt.plot(x1[:,0], x1[:,1],'ro')
plt.plot(x2[:,0], x2[:,1],'g*')
plt.plot(x3[:,0], x3[:,1],'bo');
plt.plot(x4[:,0], x4[:,1],'b*')
x = np.linspace(0.9,1.7,100)
y = -(w[0,2]+w[0,0]*x)/w[0,1]
plt.plot(x,y,'c')
plt.legend(['$A_f$','$A_{pf}$','$A_f\ Test$','$A_{pf}\ Test$','boundary'])
plt.xlabel('触角长度');plt.ylabel('翅膀长度');plt.title('感知器算法-蠓虫分类问题')
```
